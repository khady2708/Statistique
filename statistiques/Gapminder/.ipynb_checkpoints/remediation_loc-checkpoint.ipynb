# Remediation `loc` et `iloc`

Les deux sélecteurs les plus importants du monde 

```python
import pandas as pd

df = pd.read_csv(
    'data_gapminder/income.csv', 
    index_col='country'
)
```

## Sélecteur `iloc`

**Ressources** : 
- https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iloc.html

```python
df.head()
```

```python
# Ligne #0, col #0    = afghanistan en 1800
df.iloc[0, 0]
```

```python
# Récupérer l'ensemble de la ligne #2
df.iloc[2, :]
```

```python
# Raccourci de notation pour df.iloc[2, :]
# ( syntactic sugar )
df.iloc[2]
```

```python
# Récupérer toutes les colonnes pour les lignes de 
# Albanie jusqu'à 'United Arab Emirates' inclus
df.iloc[2:5, :]
```

```python
# Récupérer de la ligne #2 jusqu'à la fin pour les annees 1800, 1801 et 1802
df.iloc[2:, 0:3]
```

```python
# Récupérer toutes la ligne 0 pour les 50 dernières années
df.iloc[0, -50:]
```

## Sélecteur `loc`

Différence avec `iloc` :
- `iloc` ne veut veut que les index des lignes et colones (nbres entiers)
- `loc` veut le label (= le nom) de la ligne ou colonne

```python
# Récupérer toutes les colonnes pour le pays Afghanistan
df.loc['Afghanistan', :]
```

```python
# Récupérer toutes les colonnes pour les pays Afghanistan jusqu'à Andorra
df.loc['Afghanistan':'Andorra', :]
```

```python
# Récupérer les infos pour 'Angola' et 'Albania' pour les années 1800 à 1805
# df.loc['Angola':'Albania', :'1805']
df.loc['Angola':'Albania', '1800':'1805']
```

```python
# Récupérer les vraies infos pour tous les pays
# (sans projection future)
df.loc[:, :'2024']
```
